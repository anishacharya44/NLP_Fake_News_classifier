```python
# Text pre-processing

def text_transformation(text):
    
    text = text.lower()
    text = re.sub(r'http\S+', ' ', text) # remove urls
    text = re.sub(r'[^\w\s]', '', text) # remove punctuations and special chars
    text = text.replace('  ',' ') # remove extra whitespaces
    
    return text
```

```python
wnl = WordNetLemmatizer()
ps = PorterStemmer()
corpus = []

for i in np.arange(len(df2)):
    print(i)
    #print(df['all_text'][i])
    news = text_transformation(df2['all_text'][i])
    #print()
    #print(news)
    news = news.split()
    #news = [wnl.lemmatize(word) for word in news if word not in stopwords.words('english')]
    news = [ps.stem(word) for word in news if word not in stopwords.words('english')]
    #print()
    #print(news)
    news = " ".join(news)
    #print()
    #print(news)
    corpus.append(news)
    #print()
    #print(corpus)
```

```python
x = np.array(corpus)
y = df2['is_fake']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=56)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

x_train = tfidf.fit_transform(x_train).toarray()
x_test = tfidf.transform(x_test).toarray()
```

```python
x_train.shape
```

```python
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
#from mlxtend.plotting import plot_confusion_matrix
```

```python
gn = GaussianNB()
bn = BernoulliNB()

models = [gn, bn]

for m in models:
    m.fit(x_train, y_train)
    print(m, ':', m.score(x_train, y_train).round(4), m.score(x_test, y_test).round(4))
```

```python
for m in models:
    print(m)
    m.fit(x_train, y_train)
    y_test_pred = m.predict(x_test)
    print('Test Classification report:')
    print(classification_report(y_test,y_test_pred))
    y_train_pred = m.predict(x_train)
    print('Train Classification report:')
    print(classification_report(y_train, y_train_pred))
    print()
```
